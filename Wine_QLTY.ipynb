# Feature Selection

```python
# Create Classification version of target variable
wine['goodquality'] = [1 if x >= 7 else 0 for x in wine['quality']]# Separate feature variables and target variable
X = wine.drop(['quality','goodquality'], axis = 1)
Y = wine['goodquality']
```

```python
# See proportion of good vs bad wines
wine['goodquality'].value_counts()
```

# Feature Importance

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

from sklearn.ensemble import ExtraTreesClassifier
classifiern = ExtraTreesClassifier()
classifiern.fit(X,Y)
score = classifiern.feature_importances_
print(score)
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=7)
```

# LogisticRegression:

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix
log_accuracy = accuracy_score(Y_test, Y_pred)
print("Logistic Regression Accuracy Score:",(log_accuracy))
```

```python
confusion_mat = confusion_matrix(Y_test,Y_pred)
print(confusion_mat)
```

# Using KNN:

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
knn_accuracy = accuracy_score(Y_test, y_pred)
print("K-Nearest Neighbors Accuracy Score:", knn_accuracy)
```

# Using SVC:

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,Y_train)
pred_y = model.predict(X_test)

from sklearn.metrics import accuracy_score
svc_accuracy = accuracy_score(Y_test, pred_y)
print("Support Vector Classifier Accuracy Score:", svc_accuracy)
```

# Using Decision Tree:

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy',random_state=7)
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
dt_accuracy = accuracy_score(Y_test, y_pred)
print("Decision Tree Accuracy Score:", dt_accuracy)
```

# Using GaussianNB:

```python
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(X_train,Y_train)
y_pred3 = model3.predict(X_test)

from sklearn.metrics import accuracy_score
gnb_accuracy = accuracy_score(Y_test, y_pred3)
print("Gaussian Naive Bayes Accuracy Score:", gnb_accuracy)
```

# Using Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(random_state=1)
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)

from sklearn.metrics import accuracy_score
rf_accuracy = accuracy_score(Y_test, y_pred2)
print("Random Forest Accuracy Score:", rf_accuracy)
```

# Using Xgboost:

```python
import xgboost as xgb
model5 = xgb.XGBClassifier(random_state=1)
model5.fit(X_train, Y_train)
y_pred5 = model5.predict(X_test)

from sklearn.metrics import accuracy_score
xgb_accuracy = accuracy_score(Y_test, y_pred5)
print("XGBoost Accuracy Score:", xgb_accuracy)
```

```python
# Conclude the best one based on accuracy score
accuracies = {
    'Logistic Regression': log_accuracy,
    'K-Nearest Neighbors': knn_accuracy,
    'Support Vector Classifier': svc_accuracy,
    'Decision Tree': dt_accuracy,
    'Gaussian Naive Bayes': gnb_accuracy,
    'Random Forest': rf_accuracy,
    'XGBoost': xgb_accuracy
}

results_df = pd.DataFrame(list(accuracies.items()), columns=['Model', 'Score'])
# Sort the DataFrame by the accuracy scores in descending order
results_df = results_df.sort_values(by='Score', ascending=False)
results_df = results_df.set_index('Score')
print(results_df)
```

```python
best_model = max(accuracies, key=accuracies.get)
best_accuracy = accuracies[best_model]

print("\nThe best model is:", best_model)
print("With an accuracy score of:", best_accuracy)
```
